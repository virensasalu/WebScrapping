#### 3-e. For the first movie, extract the part of the url following the string “https://www.themoviedb.org/” using the appropriate HTML tags to extract this portion on the object created in part 3a (do not use built-in string methods)

```python
def movie_link(content_of_tmdb):
    links = content_of_tmdb.find_all('a', class_='image')
    tagssss = []
    for hre in links:
        final_link = hre['href']
        tagssss.append(final_link)
    return tagssss
```

#### 4-d Genres of all the movies on the page as a Python list

```python
def genre_ofall_movies(content_of_tmdb):
    first_movie_link = movie_link(content_of_tmdb)
    all_movie_genres = []
    for emlemnets in first_movie_link:
        base_url = 'https://www.themoviedb.org'
        movie_url = f'{base_url}{emlemnets}'
        responses = requests.get(movie_url, headers=needed_header)
        soup_individual = BeautifulSoup(responses.content, 'html.parser')
        genres_spans = soup_individual.find_all('span', class_='genres')
        movie_genres_combined = []
        
        for genres_span in genres_spans:
            a = genres_span.find_all('a')
            genres = [a.get_text() for a in a]
            combined_genres = ', '.join(genres)
            movie_genres_combined.append(combined_genres)
        all_movie_genres.extend(movie_genres_combined)
    return all_movie_genres
```

```python
print(genre_ofall_movies(content_of_tmdb))
```

#### 4-e.Cast of all the movies on the page as a Python list

```python
def cast_ofall_movies(content_of_tmdb):
    first_movie_link = movie_link(content_of_tmdb)
    all_movie_cast = []
    
    for emlemnets in first_movie_link:
        base_url = 'https://www.themoviedb.org'
        movie_url = f'{base_url}{emlemnets}'
        response = requests.get(movie_url, headers=needed_header)
        soup_individual = BeautifulSoup(response.content, 'html.parser')

        cast_elements = soup_individual.find_all('li', class_='card')
        movie_cast = []

        for cast_element in cast_elements:
            a = cast_element.find('a')
            if a and a.img and 'alt' in a.img.attrs:
                actor_name = a.img['alt']
                movie_cast.append(actor_name)

        all_movie_cast.append(', '.join(movie_cast))
    
    return all_movie_cast
#The presence of '蒋雯丽' so many problems (TypeError: 'NoneType' object is not subscriptable). 
#I found a way to skip the if an 'a' tage was missing.
```

## Task 5: Write an user defined function that returns a pandas data frame with following data: ( 5 marks )

```python
def themoviedataframe(content_of_tmdb):
    XTitles = list_of_movies(content_of_tmdb)
    XRatings = rating_of_movies(content_of_tmdb)
    XGenres = genre_ofall_movies(content_of_tmdb)
    XCasts = cast_ofall_movies(content_of_tmdb)

    data = {'Titles' : XTitles,
            'Ratings': XRatings,
            'Genres' : XGenres,
            'Casts'  : XCasts
            }
    The_movie_database = pd.DataFrame(data)
    return The_movie_database
```
